### 1. Escreva uma função chamada vogal que receba uma letra e verifique se a letra é uma vogal, retornando o valor True nesse caso, ou o valor False caso contrário. 

### Escreva também um programa que leia uma frase e, usando a função vogal criada, imprima a quantidade de vogais existentes na frase lida.

```python
def vogal(letra):
  vogais = 'AEIOU'
  if vogais.find(letra.upper()) != -1:
    return True
  return False

frase = input('Frase: ')
cont = 0
for i in frase:
  if vogal(i):
    cont += 1

print(f'Qtd. de vogais: {cont}')
```

### 2. Escreva uma função chamada primo para determinar se um número é primo ou não. A função deve receber um número inteiro, retornar True se o número for primo, ou False caso contrário.

### Escreva também um programa que, usando a função primo criada, exiba os números primos entre 1 e 100.

```python
def primo(num):
  resultado = True
  if num <= 2:
    resultado = False
  for i in range(2, num):
    if num%i == 0:
      resultado = False
  return resultado

for i in range(1, 101):
  if primo(i):
    print(i, end=' ')
```

### 3. Escreva um programa que leia N números inteiros (máximo 20) e armazene em um vetor X. Calcule a média dos elementos do vetor e informe quantos elementos estão abaixo da média do conjunto.
Crie as seguintes funções:
* Uma função que faça a leitura dos elementos de um vetor de inteiros. Essa função
deve receber como parâmetro o número de elementos do vetor e deve retornar o
vetor preenchido.
* Uma função que calcule a média dos elementos de um vetor. Essa função deve
receber como parâmetro um vetor e retornar a média dos elementos dele.
* Uma função que receba um vetor e um número, e retorne quantos elementos do
vetor estão abaixo desse número.

```python
def leitura_vetor(num):
  vetor = [None]*num
  for i in range(num):
    vetor[i] = int(input(f'V[{i}]: '))
  return vetor
def media(vetor):
  n = len(vetor)
  soma = 0
  for i in vetor:
    soma += i
  return soma/n
def abaixo_n(vetor, num):
  cont = 0
  for i in vetor:
    if i < num:
      cont += 1
  return cont
# Testes
v = leitura_vetor(3)
med = media(v)
menor_n = abaixo_n(v, 7)

print(f'\nV: {v}\n\
Média: {med:.2f}\n\
Abaixo de 7: {menor_n}')
```

### 4. Faça um programa que leia duas matrizes de inteiros e gere uma terceira matriz que corresponderá a soma das duas matrizes lidas. A ordem das matrizes também será lida.
O programa deverá conter as seguintes funções:
* Uma função para gerar e ler uma matriz, sendo passados como parâmetros a ordem
da matriz.
* Uma função para exibir uma matriz em sua representação clássica (linhas e colunas).
* Uma função para somar duas matrizes.

```python
def leitura_matriz(lin, col):
  mat = [[None]*col for i in range(lin)]
  for i in range(lin):
    for j in range(col):
      mat[i][j] = int(input(f'M[{i}][{j}]: '))
  return mat

def soma_matriz(m1, m2):
  lin = len(m1)
  col = len(m1[0])
  mat_soma = [[None]*col for i in range(lin)]
  for i in range(lin):
    for j in range(col):
      mat_soma[i][j] = m1[i][j] + m2[i][j]
  return mat_soma

def printMat(mat):
  lin = len(mat)
  col = len(mat[0])
  for i in range(lin):
    for j in range(col):
      print(f'{mat[i][j]}', end=' ')
    print()

# Teste
lin = int(input('Linha: '))
col = int(input('Coluna: '))

print('\nMatriz 1:')
m1 = leitura_matriz(lin, col)
print('\nMatriz 2:')
m2 = leitura_matriz(lin, col)

m_soma = soma_matriz(m1, m2)

print('\nMatriz 1:')
printMat(m1)
print('\nMatriz 2:')
printMat(m2)
print('\nMatriz Soma:')
printMat(m_soma)
```

### 5. Escreva um programa que calcule o valor do cosseno de X através de 20 termos da série abaixo:
Captura de Tela 2022-11-28 às 11.44.15 AM.png
     
Observações:
* O valor de X será lido.
* Escreva uma função para o cálculo do fatorial e outra para o cálculo da potência.
* Compare o resultado do seu programa com o resultado da função:
math.cos(numero): retorna o cosseno do número em radiano (da biblioteca math).

```python
from math import cos

def fatorial(num):
  fat = 1
  for i in range(2, num + 1):
    fat *= i
  return fat

def potencia(num, expo):
  for i in range(1, expo):
    num *= num
  return num

x = float(input('X: '))
y = 1
for i in range(1, 21):
  if i%2 != 0:
    y -= potencia(x, i*2)/fatorial(i*2)
  else:
    y += potencia(x, i*2)/fatorial(i*2)

print(f'RESULTADO = {y}')
print(f'COS (FUNC. MATH) = {cos(x)}')
```
